Q1] What is hypothesis testing in statistics?  
Ans] Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data.  
Hypothesis: A claim or assumption about a population parameter (like mean, proportion, variance).  
Null Hypothesis (H₀): The default assumption (no effect, no difference).  
Example: “The average test score = 70.”  

Alternative Hypothesis (H₁ or Ha): What we want to prove (there is an effect/difference).  
Example: “The average test score ≠ 70.”  

Test Statistic: A number calculated from sample data (like Z, t, χ², F) used to decide between H₀ and H₁.  

Significance Level (α): The threshold probability of making a wrong decision (commonly 0.05 or 5%).  
Example: α = 0.05 means we are okay with a 5% chance of wrongly rejecting H₀.  

p-value: Probability of observing the sample result (or more extreme) if H₀ is true.  
If p-value ≤ α → reject H₀ (evidence supports H₁).  
If p-value > α → fail to reject H₀ (not enough evidence).  

Steps in Hypothesis Testing  
State hypotheses → H₀ and H₁.  
Choose significance level (α) (e.g., 0.05).  
Collect data & compute test statistic.  
Find p-value or critical value.  
Make decision: Reject or fail to reject H₀.  
Conclusion: Interpret in the real-world context.  

2] What is the null hypothesis, and how does it differ from the alternative hypothesis?  
Ans] Null Hypothesis (H₀)  
The default assumption or starting point.  
It usually states that there is no effect, no difference, or no relationship in the population.  
Example:  
A new drug has no difference in effectiveness compared to the old drug.  
The average exam score = 70.  
A coin is fair (p = 0.5).  

Alternative Hypothesis (H₁ or Ha)  
The claim we want to test or find evidence for.  
It suggests there is an effect, a difference, or a relationship.  
Example:  
The new drug works better than the old drug.  
The average exam score ≠ 70.  
The coin is not fair (p ≠ 0.5).  

Difference between Null and Alternative Hypothesis  
Definition  
H₀: Assumes no effect, no difference, or status quo.  
H₁: Assumes there is an effect, difference, or change.  

Starting Assumption  
H₀: Always assumed true at the beginning of testing.  
H₁: Considered only if there’s strong evidence against H₀.  

Purpose  
H₀: To act as a baseline claim to test against.  
H₁: To represent the researcher’s or experimenter’s claim.  

Decision Rule  
H₀: Can only be “rejected” or “not rejected.”  
H₁: Is accepted when H₀ is rejected.  

Nature of Statement  
H₀: Equality ( =, ≤, ≥ ) is always included in H₀.  
H₁: Inequality ( ≠, <, > ) is used in H₁.  

Example (Exam scores, mean = 70)  
H₀: μ = 70 (mean score is 70).  
H₁: μ ≠ 70 (mean score is not 70).  

Analogy  
H₀: “Innocent until proven guilty.”  
H₁: “Guilty” (only concluded if strong evidence exists).  

Q3] Explain the significance level in hypothesis testing and its role in deciding the outcome of a test?  
Ans] The significance level, usually denoted as α (alpha), is the threshold probability for deciding whether to reject the null hypothesis (H₀).  
It represents the maximum risk you are willing to take of making a Type I error (rejecting H₀ when it is actually true).  
Role in Hypothesis Testing  
Set α (before collecting data). Example: α = 0.05.  
Collect sample & calculate test statistic (Z, t, χ², etc.).  
Compute p-value (probability of getting the observed result if H₀ is true).  
Decision rule:  
If p-value ≤ α → Reject H₀ (evidence supports H₁).  
If p-value > α → Fail to reject H₀ (not enough evidence).  

Example
Suppose we are testing if a coin is fair (H₀: p = 0.5).  
We toss the coin 100 times and get 65 heads.  
Test gives a p-value = 0.03.  
If α = 0.05 → 0.03 < 0.05 → Reject H₀ → evidence coin is not fair.  
If α = 0.01 → 0.03 > 0.01 → Fail to reject H₀ → not enough evidence.  
This shows how α controls the decision.  

Q4] What are Type I and Type II errors? Give examples of each?  
Ans] Type I Error (False Positive)  
Definition: Rejecting the null hypothesis (H₀) when it is actually true.  
In simple words: Saying there is an effect or difference when in reality, there isn’t.  
Probability of Type I Error = α (significance level).  

Example:  
Medical test: A healthy patient (H₀: patient is healthy) is diagnosed as sick.  
Court analogy: Declaring an innocent person guilty.  
Coin toss: Declaring a coin unfair when it is actually fair.  

Type II Error (False Negative)  
Definition: Failing to reject the null hypothesis (H₀) when it is actually false.  
In simple words: Missing a real effect or difference.  
Probability of Type II Error = β.  
Power of a test = 1 − β (ability to detect a true effect).  

Example:  
Medical test: A sick patient (H₀ is false) is diagnosed as healthy.  
Court analogy: Declaring a guilty person innocent.  
Coin toss: Declaring a biased coin fair when it is not.  

Q5] What is the difference between a Z-test and a T-test? Explain when to use each.  
Ans] Z-Test  
Definition: A hypothesis test that uses the standard normal distribution (Z-distribution).  
When to use:  
Sample size is large (n ≥ 30).  
Population standard deviation (σ) is known.  
Data should be approximately normally distributed.  
Example:  
A company claims that the mean weight of chips packets is 50g. You take 100 samples (large sample), and the population standard deviation (σ) is known = 2g. → Use Z-test.  

T-Test  
Definition: A hypothesis test that uses the t-distribution (heavier tails than normal).  
When to use:  
Sample size is small (n < 30).  
Population standard deviation (σ) is unknown (we estimate using sample standard deviation, s).  
Data should be approximately normally distributed.  
Types of T-tests:  
One-sample T-test → Compare sample mean to population mean.  
Independent two-sample T-test → Compare means of two groups.  
Paired T-test → Compare means before vs after (same group).  
Example:  
Testing whether a new teaching method improves test scores. Only 20 students are tested, and σ is unknown → Use T-test.  

Sample Size  
Z-test: Used for large samples (n ≥ 30).  
T-test: Used for small samples (n < 30).  

Population Standard Deviation (σ)  
Z-test: Known.  
T-test: Unknown (estimated from sample).  

Distribution Used  
Z-test: Standard normal distribution (Z-distribution).  
T-test: Student’s t-distribution (wider tails).  

Shape of Curve  
Z-test: Curve is narrower (less variability).  
T-test: Curve is wider (more variability, accounts for small n).  

Applicability  
Z-test: When data variance is less uncertain (large n, σ known).  
T-test: When data variance is more uncertain (small n, σ unknown).  

Use Cases  
Z-test: Comparing sample mean to population mean when σ is known, proportion tests, large sample cases.  
T-test: Comparing means of small samples, paired samples, or two groups when σ is unknown.  

Examples  
Z-test: Testing if the average weight of 1000 apples is 150g (σ known).  
T-test: Testing if 15 students’ exam scores differ from the national average (σ unknown).  

Q6] Write a Python program to generate a binomial distribution with n=10 and
p=0.5, then plot its histogram.
(Include your Python code and output in the code box below.)
Hint: Generate random number using random function.

```python
import numpy as np
import matplotlib.pyplot as plt

n = 10   
p = 0.5     
size = 1000 

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), align='left', rwidth=0.8, edgecolor='black')
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of successes")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

7] : Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
(Include your Python code and output in the code box below.)

```python
import numpy as np
from scipy.stats import norm

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

mu_0 = 50

sigma = 0.5

sample_mean = np.mean(sample_data)
n = len(sample_data)

z_stat = (sample_mean - mu_0) / (sigma / np.sqrt(n))

p_value = 2 * (1 - norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("p-value:", p_value)
```

8] Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(42)
data = np.random.normal(loc=50, scale=5, size=100)  

sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  
n = len(data)
std_error = sample_std / np.sqrt(n)

confidence = 0.95
alpha = 1 - confidence
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
margin_of_error = t_critical * std_error
ci_lower, ci_upper = sample_mean - margin_of_error, sample_mean + margin_of_error

plt.hist(data, bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci_lower, color='green', linestyle='--', label=f"95% CI Lower = {ci_lower:.2f}")
plt.axvline(ci_upper, color='green', linestyle='--', label=f"95% CI Upper = {ci_upper:.2f}")
plt.title("Normal Distribution Simulation with 95% CI")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()

print("Sample Mean:", sample_mean)
print("95% Confidence Interval:", (ci_lower, ci_upper))
```

9] Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.


```python
import numpy as np
import matplotlib.pyplot as plt

def z_score_analysis(data):
    
    data = np.array(data)
    
    mean = np.mean(data)
    std = np.std(data)
    
    z_scores = (data - mean) / std
    
    print("Mean of data:", mean)
    print("Standard deviation of data:", std)
    print("Z-scores:", z_scores)
    
    plt.hist(z_scores, bins=10, color='skyblue', edgecolor='black')
    plt.title("Histogram of Z-scores")
    plt.xlabel("Z-score")
    plt.ylabel("Frequency")
    plt.show()
    
    return z_scores

sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.0, 49.6]
z_score_analysis(sample_data)
```

Definition: A Z-score tells you how many standard deviations a data point is from the mean of the dataset.  
Formula:  
𝑍=(𝑋−𝜇)/𝜎  
 
Where:  
X = individual data point  
μ = mean of the dataset  
σ = standard deviation  

Interpretation:  
Z = 0 → the value is exactly at the mean.  
Z = 1 → the value is 1 standard deviation above the mean.  
Z = -2 → the value is 2 standard deviations below the mean.  
By standardizing the data using Z-scores, we can compare values from different datasets on the same scale, even if the original units are different. The histogram shows the distribution of the data in terms of standard deviations.
