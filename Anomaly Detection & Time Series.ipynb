 Q1] What is Anomaly Detection? Explain its types (point, contextual, and
collective anomalies) with examples.  
Ans] Anomaly Detection (also called outlier detection) is the process of identifying data points, events, or patterns that deviate significantly from normal behavior.  

These anomalies usually indicate critical incidents, such as:  
Fraudulent transactions  
Network intrusions  
Equipment failures  
Unusual system behavior  

1. Point Anomalies  
A single data point that is far away from the rest of the data.  
Example:  
Credit card transaction of ₹5,00,000 when a user usually spends ₹1,000–₹5,000  
Temperature sensor suddenly reading 120°C, when typical readings are 30–40°C  

2. Contextual Anomalies  
A data point that is unusual in a specific context, such as time, location, or situation.  
Here, the same value may be normal in one context but abnormal in another.  
Example:  
Temperature reading of 30°C  
Normal in summer  
Anomaly in winter  
Website traffic spike at 2 AM  
Normal during a sale  
Anomaly on a regular day  

3. Collective Anomalies  
A group of related data points that together form an anomaly, even if individual points may not be unusual.  
Example:  
A sequence of network packets that looks like a DDoS attack  
A batch of machines showing gradually dropping productivity  
Continuous drop in stock price over 10 days, even if each single-day drop is small  
Individually, each point looks normal, but collectively they indicate a pattern of concern.  



2] Compare Isolation Forest, DBSCAN, and Local Outlier Factor in terms of
their approach and suitable use cases.  
Ans] 
1. Isolation Forest  
Approach  
Based on randomly partitioning data using decision trees.  
Anomalies are isolated quickly because they require fewer splits.  
Normal points require more splits to be isolated.  
Produces an anomaly score for each point.  
Best For / Suitable Use Cases  
High-dimensional datasets  
Large datasets (scalable, fast)  
Mixed-type data (numerical + categorical with preprocessing)  
Problems where global outliers need to be found  

2. DBSCAN (Density-Based Spatial Clustering of Applications with Noise)  
Approach  
Groups data into clusters based on density.  
Points in low-density areas are treated as outliers/noise.  
Requires two parameters: ε (epsilon) and minPts.  
Does not assume any fixed number of clusters.  
Best For / Suitable Use Cases  
Detecting clusters of arbitrary shape  
Data with well-defined dense regions and sparse noise  
Spatial data and 2D/3D datasets  

3. LOF (Local Outlier Factor)  
Approach  
Compares the local density of a point to its neighbors.  
Points with significantly lower local density are outliers.  
Values > 1 indicate outliers.  
Best For / Suitable Use Cases  
Datasets where anomaly depends on local neighborhood structure  
Detecting contextual anomalies  
Non-uniform density data  



3] What are the key components of a Time Series? Explain each with one
example.  
Ans] Key Components of a Time Series  
A time series is typically made up of four major components:  
Trend (T)  
Seasonality (S)  
Cyclic Component (C)  
Irregular / Noise Component (I)  

1. Trend (T)  
The overall long-term movement of the data, either upward, downward, or stable.  
It shows how values change over years or months, ignoring short-term fluctuations.  
Example  
House prices increasing steadily over 10 years.  
Trend → upward over time.  

2. Seasonality (S)  
Regular, repeating patterns that occur at fixed intervals, such as:  
Daily  
Weekly  
Monthly  
Yearly  
Seasonality is caused by predictable factors like weather, festivals, and holidays.  
Example  
Ice cream sales peak every summer and drop in winter.  
Repeating pattern every year → seasonal.  

3. Cyclic Component (C)  
Long-term up-and-down movements that are not regular like seasonality.  
Cycles usually occur due to:  
Business cycles  
Economic conditions  
Market trends  
They last multiple years and are not fixed in frequency.  
Example  
Economic recession every 7–10 years affecting GDP.  
This rise–fall pattern is a cycle, but the timing is not fixed.  

4. Irregular / Noise Component (I)  
Random variations or shocks in the data that:  
Are unpredictable  
Do not repeat  
Occur due to sudden events  
Example  
A sudden drop in stock price due to unexpected news.  
This jump is noise, not a trend or seasonality.  

4] Define Stationary in time series. How can you test and transform a
non-stationary series into a stationary one?  
Ans] A time series is stationary if its statistical properties do not change over time. 
1. Differencing  
Subtract the current value from the previous value.	​  
Removes trend  
First differencing is usually enough  
If needed, second differencing can be applied  
Example:  
Stock prices series → difference to remove growth trend.  

2. Seasonal Differencing  
Used when series has seasonality.  
Where s = seasonal period  
(e.g., 12 for monthly data)  
Example:  
Monthly sales with yearly seasonality → apply 12-lag differencing.  

3. Log Transformation  
Used to stabilize variance.  
Helpful when data has exponential growth.  
Example:  
GDP or population growth time series.  

4. Square Root / Box-Cox Transformation  
Alternative to log when data contains zeros.  
Box-Cox reduces variance and makes data more Gaussian.  

5. Detrending  
Removing trend explicitly using:  
Regression models  
Moving averages  
Example:  
Fit a line to a trending series → subtract fitted values.  

6. Remove Seasonality  
Methods:  
Seasonal decomposition (STL)  
Subtract seasonal component  
Useful for strong yearly/monthly patterns.   


5] Differentiate between AR, MA, ARIMA, SARIMA, and SARIMAX models in
terms of structure and application.  
Ans] 
1. AR (Auto-Regressive) Model  
Structure  
Uses past values of the series to predict the future.  
Model notation: AR(p)  
When to Use  
When the time series shows correlation with its own past values.  
Suitable for stationary data.  
Example Application  
Predicting next day temperature using previous days’ temperatures.  

2. MA (Moving Average) Model  
Structure  
Uses past forecast errors (shocks) to explain the current value.  
Model notation: MA(q)  	​
When to Use  
When random shocks/errors influence the series.  
Good for stationary data.  
Example Application  
Modeling stock returns influenced by unexpected market shocks.  

3. ARIMA (Auto-Regressive Integrated Moving Average)  
Structure  
Combines AR, I (differencing), and MA components.  
Model notation: ARIMA(p, d, q)  
p = AR order  
d = differencing (to make series stationary)  
q = MA order  
When to Use  
For non-stationary time series with trend (but no seasonality).  
After applying differencing to remove trend.  
Example Application  
Forecasting monthly sales with trend but no seasonal cycle.  

4. SARIMA (Seasonal ARIMA)  
Structure  
Extends ARIMA by adding seasonal components.  
Model notation:  
SARIMA(p, d, q) × (P, D, Q, s)  
P, D, Q = seasonal AR, differencing, MA  
s = seasonal period (e.g., 12 for monthly data)  
When to Use  
When time series has both trend and strong seasonality.  
Handles repeating patterns such as quarterly or monthly cycles.  
Example Application  
Forecasting monthly electricity demand with yearly seasonality.  

5. SARIMAX (Seasonal ARIMA with Exogenous Variables)  
Structure  
SARIMA + extra external inputs (X-variables).  
Model:  
SARIMAX(p, d, q)(P, D, Q, s) + X  
Allows inclusion of external regressors.  
When to Use  
Seasonal data influenced by outside factors.  
When external predictors improve accuracy.  
Example Application  
Forecasting sales using:  
seasonal patterns  
promotions/discounts  
weather conditions  
holidays

7] Apply Isolation Forest on a numerical dataset (e.g., NYC Taxi Fare) to
detect anomalies. Visualize the anomalies on a 2D scatter plot.  

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

# Step 1: Generate synthetic dataset
np.random.seed(42)
trip_distance = np.random.normal(5, 2, 200)   # average trip ~5 miles
fare_amount = trip_distance * 2.5 + np.random.normal(0, 2, 200)  # fare ~ $2.5 per mile

# Add anomalies (unusual trips)
trip_distance_anom = np.random.uniform(15, 25, 10)   # very long trips
fare_amount_anom = np.random.uniform(50, 100, 10)    # unusually high fares

trip_distance = np.concatenate([trip_distance, trip_distance_anom])
fare_amount = np.concatenate([fare_amount, fare_amount_anom])

X = np.column_stack((trip_distance, fare_amount))

# Step 2: Fit Isolation Forest
clf = IsolationForest(contamination=0.05, random_state=42)
clf.fit(X)

# Step 3: Predict anomalies
labels = clf.predict(X)

# Step 4: Plot scatter plot
plt.figure(figsize=(8,6))
plt.scatter(X[labels==1, 0], X[labels==1, 1], c='blue', label='Normal')
plt.scatter(X[labels==-1, 0], X[labels==-1, 1], c='red', label='Anomaly')
plt.xlabel('Trip Distance (miles)')
plt.ylabel('Fare Amount ($)')
plt.title('Isolation Forest Anomaly Detection on Taxi Fare Data')
plt.legend()
plt.show()
```

8] Train a SARIMA model on the monthly airline passengers dataset.
Forecast the next 12 months and visualize the results.

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Step 1: Load dataset (AirPassengers)
url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv'
data = pd.read_csv(url, parse_dates=['Month'], index_col='Month')

# Step 2: Fit SARIMA model
# (p,d,q) = (1,1,1), seasonal (P,D,Q,12) = (1,1,1,12)
model = SARIMAX(data['Passengers'], order=(1,1,1), seasonal_order=(1,1,1,12))
results = model.fit(disp=False)

# Step 3: Forecast next 12 months
forecast = results.get_forecast(steps=12)
forecast_ci = forecast.conf_int()

# Step 4: Plot results
plt.figure(figsize=(10,6))
plt.plot(data.index, data['Passengers'], label='Observed')
plt.plot(forecast.predicted_mean.index, forecast.predicted_mean, 
         label='Forecast', color='red')
plt.fill_between(forecast_ci.index,
                 forecast_ci.iloc[:,0],
                 forecast_ci.iloc[:,1], color='pink', alpha=0.3)
plt.xlabel('Date')
plt.ylabel('Number of Passengers')
plt.title('SARIMA Forecast of Airline Passengers')
plt.legend()
plt.show()
```

9] Apply Local Outlier Factor (LOF) on any numerical dataset to detect
anomalies and visualize them using matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import LocalOutlierFactor

# Step 1: Generate synthetic dataset
np.random.seed(42)
# Normal points clustered around (2,2) and (-2,-2)
X_inliers = 0.3 * np.random.randn(200, 2)
X_inliers = np.r_[X_inliers + 2, X_inliers - 2]

# Add some anomalies spread out randomly
X_outliers = np.random.uniform(low=-6, high=6, size=(20, 2))

# Combine into one dataset
X = np.r_[X_inliers, X_outliers]

# Step 2: Fit LOF model
lof = LocalOutlierFactor(n_neighbors=20, contamination=0.1)
y_pred = lof.fit_predict(X)

# Step 3: Separate normal and anomalies
X_normal = X[y_pred == 1]
X_anomaly = X[y_pred == -1]

# Step 4: Plot scatter plot
plt.figure(figsize=(8,6))
plt.scatter(X_normal[:, 0], X_normal[:, 1], c='blue', label='Normal')
plt.scatter(X_anomaly[:, 0], X_anomaly[:, 1], c='red', label='Anomaly')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Local Outlier Factor (LOF) Anomaly Detection')
plt.legend()
plt.show()
```

10] You are working as a data scientist for a power grid monitoring company.  
Your goal is to forecast energy demand and also detect abnormal spikes or drops in  
real-time consumption data collected every 15 minutes. The dataset includes features
like timestamp, region, weather conditions, and energy usage.  
Explain your real-time data science workflow:  
● How would you detect anomalies in this streaming data (Isolation Forest / LOF /
DBSCAN)?  
● Which time series model would you use for short-term forecasting (ARIMA /
SARIMA / SARIMAX)?  
● How would you validate and monitor the performance over time?  
● How would this solution help business decisions or operations?  

Ans]
1. Detecting anomalies in streaming data  
Isolation Forest:  
Works well for high-dimensional data (energy usage + weather + region).   
Quickly isolates abnormal spikes/drops in consumption relative to historical patterns.  
Local Outlier Factor (LOF):  
Useful when consumption varies by region and local density matters.  
Detects anomalies relative to neighboring points (e.g., one region suddenly deviates from similar regions).  
DBSCAN (Density-Based Clustering):  
Effective for clustering streaming data and flagging points that don’t belong to any cluster.  
Good for detecting rare events like sudden outages or extreme weather-driven demand.  
In practice:  
Use Isolation Forest for fast global anomaly detection.  
Complement with LOF for local/regional anomalies.  
DBSCAN can be applied periodically to detect structural shifts in consumption patterns.  

2. Short-term forecasting model  
SARIMAX (Seasonal ARIMA with Exogenous Variables):  
Best suited because energy demand has strong seasonality (daily, weekly, yearly).  
Incorporates weather conditions (temperature, humidity) as exogenous regressors.  
Captures both short-term fluctuations and seasonal cycles.  
Example: Forecast next 12 hours (48 steps at 15-min intervals) using SARIMAX with weather features.  
  
3. Validation and performance monitoring  
Backtesting:  
Split historical data into rolling windows.  
Compare forecasts against actuals using metrics like RMSE, MAE, and MAPE.  
Real-time monitoring:  
Deploy dashboards showing forecast vs. actual demand.  
Track anomaly alerts with severity scores.  
Model drift detection:  
Continuously monitor error metrics.  
Retrain models when performance degrades (e.g., new consumption patterns due to policy changes or climate events).  

4. Business impact  
Operational efficiency:  
Forecasting demand helps optimize power generation schedules, reducing fuel costs.  
Grid stability:  
Early anomaly detection prevents blackouts by flagging sudden spikes/drops in consumption.  
Demand response programs:  
Utilities can incentivize customers to reduce usage during peak demand.  
Strategic planning:  
Long-term insights into seasonal demand help with infrastructure investment decisions.
